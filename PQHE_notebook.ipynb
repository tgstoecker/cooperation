## 2. Creation of a conf. file with this layout:  
(this is the layout of the input for our calculation)
```
b   = 0.25   # F2 population  
m1  = 50     # individuals in pool-1  
m2  = 50     # individuals in pool-2  
n   = 531    # sample size of population  
AF1 = 0.724  # estimated bulked allele frequency in pool-1  
AF2 = 0.114  # estimated bulked allele frequency in pool-2  
```

- **The following cell can be adapted to perform the analysis for each QTL.**
- **Alternatively, have multiple cells - each for a different QTL.**
- **This cell demonstratively peforms set up for the third QTL region (alternative allele comparison between mutant and WT bulk) - based on the  unfiltered_interesting_qtls_delta_snp.csv results file in our sciebo folder.**



````r
#moving into the PQHE directory
setwd('/PQHE/')

#we create an input file with our specifications
fileConn<-file("input_exp1.txt")
writeLines(c
            ("b   = 0.25   # F2 population",  
             "m1  = 143     # individuals in pool-1",  
             "m2  = 143     # individuals in pool-2",  
             "n   = 640    # sample size of population",
             "AF1 = 0.964292979454104  # estimated bulked allele frequency in pool-1",
             "AF2 = 0.364750850696872  # estimated bulked allele frequency in pool-2"),
            fileConn)
close(fileConn)
````

````r
#we run the programme with input we just created and write to an output object
result <- system("Rscript estimate_QTL_heritability.v0.4.10.R input_exp1.txt", intern=TRUE)
#we print the results on a per-line basis 
cat(paste(result, collapse='\n'))
````
